#Setup

```python
import pandas as pd
from IPython.display import display

df = pd.read_excel("https://github.com/Amanda0975/MT/raw/main/raw/Elecciones%20Consejeros%202023.xlsx")
```

```python
df=df.drop(columns=["Región", "Pacto"])
df.rename(columns={'Unnamed: 4': 'Coalición'}, inplace=True)
```

```python
#Creación columnas de Candidato ordenados según cantidad de votos por comuna.
df.sort_values(by=['Comuna', 'Votos'], ascending=[True, False], inplace=True)

# Crear un nuevo DataFrame con una fila por comuna
df1 = df.groupby('Comuna').head(1).reset_index(drop=True)

# Agregar las nuevas columnas de candidatos y rellenarlas con NaN
for i in range(1, 9):
    df1[f'Candidato{i}'] = None

# Llenar las columnas de candidatos con las coaliciones correspondientes
for _, row in df1.iterrows():
    comuna = row['Comuna']
    df_comuna = df[df['Comuna'] == comuna]
    for i, (_, candidato_row) in enumerate(df_comuna.iterrows(), 1):
        df1.at[row.name, f'Candidato{i}'] = candidato_row['Coalición']
```

```python
df1=df1.drop(columns=["Coalición", "Votos"])
```

```python
df1=df1.drop(columns=["Candidato6", "Candidato7","Candidato8"])
```

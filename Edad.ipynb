#Setup

```python
import pandas as pd
from IPython.display import display

df = pd.read_excel("https://github.com/Amanda0975/MT/raw/main/raw/estimaciones-y-proyecciones-2002-2035-comunas.xlsx")
```

```python
#Reestructuración de tabla
df=df.drop(columns=["Nombre Region", "Provincia", "Nombre Provincia", "Sexo\n1=Hombre\n2=Mujer","Poblacion 2017","Comuna"])
df=df.rename(columns={"Nombre Comuna": "Comuna"})
df=df.rename(columns={"Region": "region_id"})
```

##Database tramos por edad

```python
#Creación database por tramo de edad

#Adultez joven
df1 = df[(df['Edad'] >= 18) & (df['Edad'] <= 29)]

#Adultez media
df2 = df[(df['Edad'] >= 30) & (df['Edad'] <= 60)]
```

###Adultez joven

```python
df1=df1.groupby(['Comuna', 'region_id'])[['Poblacion 2021', 'Poblacion 2022']].sum().reset_index()
```

```python
df1=df1.rename(columns={"Poblacion 2021": "Poblacion 2021_1"})
df1=df1.rename(columns={"Poblacion 2022": "Poblacion 2022_1"})
```

```python
display(df1)
```

###Adultez media

```python
df2=df2.groupby(['Comuna', 'region_id'])[['Poblacion 2021', 'Poblacion 2022']].sum().reset_index()
```

```python
df2=df2.rename(columns={"Poblacion 2021": "Poblacion 2021_2"})
df2=df2.rename(columns={"Poblacion 2022": "Poblacion 2022_2"})
```

```python
display(df2)
```

####Merge df (Proporción)

```python
data= pd.merge(df1, df2, on=['Comuna', 'region_id'], how='right')
```

```python
#Creación columnas de sumatoria
data["SUM Población 2021"] = data["Poblacion 2021_1"] + data["Poblacion 2021_2"]
data["SUM Población 2022"] = data["Poblacion 2022_1"] + data["Poblacion 2022_2"]
```

```python
#Porcentaje
data["Poblacion 2021_1"] = data["Poblacion 2021_1"]/data["SUM Población 2021"]
data["Poblacion 2021_2"] = data["Poblacion 2021_2"]/data["SUM Población 2021"]
data["Poblacion 2022_1"] = data["Poblacion 2022_1"]/data["SUM Población 2022"]
data["Poblacion 2022_2"] = data["Poblacion 2022_2"]/data["SUM Población 2022"]
```

```python
display(data)
```

```python
data.drop(columns=["SUM Población 2021", "SUM Población 2022"])
```

```python
import pandas as pd
from IPython.display import display

df = pd.read_excel("https://github.com/Amanda0975/MT/raw/main/raw/estimaciones-y-proyecciones-2002-2035-comunas.xlsx")
```

```python
df =df.drop(columns=["Nombre Region", "Provincia", "Nombre Provincia", "Sexo\n1=Hombre\n2=Mujer", "Edad","Poblacion 2017","Comuna"])
df=df.rename(columns={"Nombre Comuna": "Comuna"})
df=df.rename(columns={"Region": "region_id"})
```

```python
df['Comuna'] = df['Comuna'].str.upper()
df1=df.drop(columns=["region_id"])
df1=df1.groupby('Comuna').sum().drop_duplicates()
df2=df[['region_id','Comuna']].copy().drop_duplicates()
df=pd.merge(df1,df2,on=['Comuna'],how='left').drop_duplicates()
```

```python
#Exportar .csv
df.to_csv('Poblacion_2021_2022.csv', index=False)
```

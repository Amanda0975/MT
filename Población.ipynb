```python
import pandas as pd
from IPython.display import display

df = pd.read_excel("https://github.com/Amanda0975/MT/raw/main/raw/estimaciones-y-proyecciones-2002-2035-comunas.xlsx")
```

```python
df_1 =df.drop(columns=["Nombre Region", "Provincia", "Nombre Provincia", "Sexo\n1=Hombre\n2=Mujer", "Edad","Poblacion 2017","Comuna"])
```

```python
df_1=df_1.rename(columns={"Nombre Comuna": "Comuna"})
```

```python
df_1.nunique()
```

```python
df_1=df_1.groupby("Comuna").sum().drop_duplicates()
```

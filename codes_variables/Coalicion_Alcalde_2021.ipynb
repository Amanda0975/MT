#Setup


```python
import pandas as pd
from IPython.display import display

df = pd.read_excel("https://github.com/Amanda0975/MT/raw/main/raw/2021_05_Alcaldes_Datos_Eleccion.xlsx")
```

```python
#Crear columna de Coalicion

#Coalicion1 como INDEPENDIENTES en derecha
def asignar_valor_coalicion1(pacto):
    if pacto in ["CHILE VAMOS", "INDEPENDIENTES"]:
        return 1
    else:
        return 0

#Coalicion1 como INDEPENDIENTES en izquierda
def asignar_valor_coalicion2(pacto):
    if pacto == "CHILE VAMOS":
        return 1
    else:
        return 0

df['Coalicion1'] = df['Pacto'].apply(asignar_valor_coalicion1)
df['Coalicion2'] = df['Pacto'].apply(asignar_valor_coalicion2)

#Filtrar solo cargo de alcalde
df1=df.loc[df['Cargo'] == 'ALCALDE']

#Rename columns
df1.rename(columns={'Nro.Región': 'region_id'}, inplace=True)
```

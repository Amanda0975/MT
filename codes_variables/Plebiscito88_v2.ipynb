#Setup

```python
import pandas as pd
from IPython.display import display

df = pd.read_excel("https://github.com/Amanda0975/MT/raw/main/raw/Plebiscito88_raw.xlsx")
```

```python
df=df.rename(columns={'comuna': 'Comuna'})
df=df.drop(columns=["region"])
```

```python
# Define los mapeos de valores
mapeo_valores = {
    "Alto Hospicio": "Iquique",
    "Alto Biobío": "Santa Bárbara",
    "Hualpén": "Talcahuano",
    "Cholchol": "Nueva Imperial",
    "Padre Hurtado": "Chillán",
    "Chillán Viejo": "Chillán"
}

# Aplica los mapeos y asigna los valores
for destino, origen in mapeo_valores.items():
    df.loc[df['Comuna'] == destino, 'NO'] = df.loc[df['Comuna'] == origen, 'NO'].values[0]
    df.loc[df['Comuna'] == destino, 'SI'] = df.loc[df['Comuna'] == origen, 'SI'].values[0]
    df.loc[df['Comuna'] == destino, 'BLANCO'] = df.loc[df['Comuna'] == origen, 'BLANCO'].values[0]
    df.loc[df['Comuna'] == destino, 'NULOS'] = df.loc[df['Comuna'] == origen, 'NULOS'].values[0]
```

```python
df=df.drop(columns=['BLANCO','NULOS']).drop_duplicates()
```

```python
#Creación columna "Total"
df['TOTAL']=df['NO']+df['SI']

#Porcentaje de votos por opción
df['NO']=df['NO']/df['TOTAL']
df['SI']=df['SI']/df['TOTAL']
```

```python
data=df[['region_id','Comuna','NO']].copy()
```

#Exportar

```python
data.to_csv('PLebiscito88.csv', index=False)
```

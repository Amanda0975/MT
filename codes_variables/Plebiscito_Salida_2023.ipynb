#Setup

```python
from IPython.display import display
import numpy as np
import pandas as pd

#Importar database desde github (Resultados eleccionarios)
df = pd.read_csv("https://github.com/cnavarreteliz/atlas-elections-data/raw/master/data_output/Chile/2023_plebiscite_polling_station.csv.gz", compression='gzip')
df_location = pd.read_csv("https://github.com/cnavarreteliz/atlas-elections-data/raw/master/data_output/Chile/2023_plebiscite_polling_station_location.csv.gz", compression='gzip')

#Merge de ambos database (Resultados eleccionarios)
data = pd.merge(df, df_location, on="polling_id")

#Importar database desde github (Polarización detallado por región y comuna)
df_ep_r = pd.read_csv("https://github.com/cnavarreteliz/atlas-elections-data/raw/master/data_curated/Chile/antagonism_2023_region_plebiscite.csv.gz", compression='gzip')
df_ep_c = pd.read_csv("https://github.com/cnavarreteliz/atlas-elections-data/raw/master/data_curated/Chile/antagonism_2023_commune_plebiscite.csv.gz", compression='gzip')

#Merge datos de polarización región
data = pd.merge(data, df_ep_r, on='region', how='left')

#Merge datos de polarización comuna
data = pd.merge(data, df_ep_c, on='commune', how='left')

#Eliminar filas con votos nulos y/o blancos
data = data[data['flag_candidates'] != 0]

#Renombrando columnas por región (r) y comunas (c)
data = data.rename(columns={
    "candidate_x": "candidate",
    "candidate_y": "candidate_r",
    "antagonism_x": "antagonism_r",
    "type_x": "type_r",
    "candidate": "candidate_c",
    "antagonism_y": "antagonism_c",
    "type_y": "type_c"
})

#Filtrar tipo de indicador en solo EP para región y comuna
data = data[(data['type_r'] == 'EP') & (data['type_c'] == 'EP')]

display (data)
```

```python
# Agrupar regiones por Zonas de Chile
condiciones = [
    data['region_id'].isin([15, 1, 2]),  # Norte Grande
    data['region_id'].isin([3, 4]),      # Norte Chico
    data['region_id'].isin([5, 13, 6, 7, 16, 8]),  # Zona Central
    data['region_id'].isin([9, 14, 10]),  # Zona Sur
    data['region_id'].isin([11, 12])     # Zona Austral
]

# Definir los valores correspondientes a cada condición
valores = [
    "Norte Grande",
    "Norte Chico",
    "Zona Central",
    "Zona Sur",
    "Zona Austral"
]

# Crear la columna 'zona' utilizando np.select
data['zona'] = np.select(condiciones, valores)

# Mostrar las primeras filas del DataFrame con la nueva columna 'zona'
display(data.head())
```

##Ratios por sector

```python
#Creación columna "rate_z" para ratios de resultados electorales por zona
data['rate_z'] = data.groupby(['zona', 'candidate'])['value'].transform('sum') / data.groupby('zona')['value'].transform('sum') * 100

# Redondear la columna 'rate_z' a dos decimales
data['rate_z'] = data['rate_z'].round(2)

# Mover 'rate_z' a la derecha de 'value'
columnas = list(data.columns)
indice_value = columnas.index('value')
indice_rate_z = columnas.index('rate_z')
columnas.insert(indice_value + 1, columnas.pop(indice_rate_z))
data = data[columnas]
```

```python
#Creación columna "rate_r" para ratios de resultados electorales por región
data['rate_r'] = data.groupby(['region', 'candidate'])['value'].transform('sum') / data.groupby('region')['value'].transform('sum') * 100

# Redondear la columna 'rate_r' a dos decimales
data['rate_r'] = data['rate_r'].round(2)

# Mover 'rate_r' a la derecha de 'rate_z'
columnas = list(data.columns)
indice_rate_z = columnas.index('rate_z')
indice_rate_r = columnas.index('rate_r')
columnas.insert(indice_rate_z + 1, columnas.pop(indice_rate_r))
data = data[columnas]
```

```python
# Crear una nueva columna 'rate_c' para ratios de resultados electorales por comuna
data['rate_c'] = data.groupby(['commune', 'candidate'])['value'].transform('sum') / data.groupby('commune')['value'].transform('sum') * 100

# Redondear la columna 'rate_r' a dos decimales
data['rate_c'] = data['rate_c'].round(2)

# Mover 'rate_r' a la derecha de 'value'
columnas = list(data.columns)
indice_rate_r = columnas.index('rate_r')
indice_rate_c = columnas.index('rate_c')
columnas.insert(indice_rate_r + 1, columnas.pop(indice_rate_c))
data = data[columnas]
```

#Solo comunas

```python
#Filtro database
df1 = data[["region_id", "commune", "candidate_c", "antagonism_c"]].copy()

#----------------------------------------------------------------------------------------------------------------
# Creamos dummies 'pol_apruebo' y 'pol_rechazo' con valores iniciales de 0
df1['pol_favor'] = 0
df1['pol_contra'] = 0

# Actualizamos los valores de las columnas según las condiciones especificadas
df1.loc[df1['candidate_c'] == 'A FAVOR', 'pol_favor'] = df1.loc[df1['candidate_c'] == 'A FAVOR', 'antagonism_c']
df1.loc[df1['candidate_c'] == 'EN CONTRA', 'pol_contra'] = df1.loc[df1['candidate_c'] == 'EN CONTRA', 'antagonism_c']

# Eliminamos las columnas 'candidate_c' y 'antagonism_c' si ya no son necesarias
# df1.drop(['candidate_c', 'antagonism_c'], axis=1, inplace=True)

# Si deseas mantener solo las columnas originales y las dos nuevas columnas
df1 = df1[['commune', 'region_id', 'pol_favor', 'pol_contra']]

#----------------------------------------------------------------------------------------------------------------
#Agrupamos en una fila por Comuna
df1 = df1.groupby('commune').agg({'pol_favor': 'max', 'pol_contra': 'max'}).reset_index()

# Reemplazamos los valores 0 por la información de la columna correspondiente
df1['pol_favor'] = df1.apply(lambda row: row['pol_contra'] if row['pol_favor'] == 0 else row['pol_favor'], axis=1)
df1['pol_contra'] = df1.apply(lambda row: row['pol_favor'] if row['pol_contra'] == 0 else row['pol_contra'], axis=1)

#-----------------------------------------------------------------------------------------------------------------
#Merge con region_id
df1 = pd.merge(df1, data[['commune', 'region_id']], on='commune', how='left')

#-----------------------------------------------------------------------------------------------------------------
#Rename comuna
df1 = df1.rename(columns={"commune": "Comuna"})

#-----------------------------------------------------------------------------------------------------------------
#Borrar duplicates
df1=df1.drop_duplicates()
```

```python
#Ratio antagonismo por comuna
df1['antagonism_2023'] = df1['pol_favor'] + df1['pol_contra']
df=df1[['region_id','Comuna','antagonism_2023']].copy().drop_duplicates()
```
